#Analysis 
1.The Bubble Plot shows driver count per city and total number rides per city is greatest for urban type and least for the rural type.So the Pyber rides are more popular  and easily available in the Urban cities(total rides per city=68.4%).In suburban cities availability of pyber is on the average side.In the rural areas pyber is limited.

2.The average fair is higher for rural areas as compared to the urban areas.For the suburban areas the average fare is in the medium range.As the total number of rides per city is increasing the average fair is decreasing.This shows that when greater distance is covered the fair would reduce.
Looking at the pie chart we notice that the percent of total fare by city type is greatest for Urban type(62.7%) which could be possible because of large number of rides.

3.Larger the number of drivers per city lesser the average fare.For around 40 drivers in a city the average fare has gone down to 25 dollars.In the Urban area percent of driver is 80.9% which is really high.As the availability of driver increases the prices goes down.


```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)
city_df
ride_df

# Combine the data into a single dataset
merged_df = pd.merge(ride_df,city_df, on="city",how='left')


# Display the data table for preview
merged_df.head()

#merged_df.dtypes


```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
#group the merged file based on city
city_grouped_data =merged_df.groupby("city")
city_grouped_data.count().head()

#calculate the total number of rides per city
total_ride_city = city_grouped_data["ride_id"].count()
total_ride_city

#calculate the average fare
average_fare_city = city_grouped_data["fare"].mean()
average_fare_city

#calculate the total number of driver per city
driver_count = city_grouped_data["driver_count"].mean()
driver_count

#find type for each city
type_value = city_grouped_data["type"].unique()
type_value

#create a new dataframe and store all values in it

new_df = pd.DataFrame({"Total number of rides":total_ride_city,
                      "average fare":average_fare_city,
                       "Driver count":driver_count,
                       "type_value":type_value})

new_df

#create new df to store type values=urban
new_df_urban  = new_df.loc[new_df["type_value"] == "Urban", :]
new_df_urban
#create new df to store type values=Rural
new_df_rural  = new_df.loc[new_df["type_value"] == "Rural", :]
new_df_rural
#create new df to store type values=Suburban
new_df_suburban  = new_df.loc[new_df["type_value"] == "Suburban", :]
new_df_suburban



#plot scatter plots for each city type assigning different colors to each
plt.scatter(x=new_df_urban["Total number of rides"], y=new_df_urban["average fare"],s= new_df_urban["Driver count"]*8, c="Coral", 
             edgecolor='black', linewidth=1, alpha=0.7,label ="Urban")

plt.scatter(x=new_df_suburban["Total number of rides"], y=new_df_suburban["average fare"],s= new_df_suburban["Driver count"]*8, c="SkyBlue", 
             edgecolor='black', linewidth=1, alpha=0.7,label = "Surburban")
plt.scatter(x=new_df_rural["Total number of rides"], y=new_df_rural["average fare"],s= new_df_rural["Driver count"]*8, c="Gold", 
             edgecolor='black', linewidth=1, alpha=0.7,label = "Rural")

# Incorporate the other graph properties
plt.grid(color='black',linewidth ='0.5')

# Create a legend
plt.xlabel("Total number of Rides(per city)")
plt.ylabel("Average Fare$")
plt.suptitle("Pyber Ride Sharing Data 2018")
plt.title("Note:circle size correlates with driver count per city")

# Incorporate a text label regarding circle size
my_legend = plt.legend(loc='upper right', title='City Types', 
                       scatterpoints=1)

for handle in my_legend.legendHandles:
    handle.set_sizes([50])


# Save Figure
plt.savefig("images/Bubbleplot.png")


plt.show()
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
#group the data based on type
type_grouped_data =merged_df.groupby("type")
type_grouped_data.count().head()

# Calculate total fare for each type
total_fare_type = type_grouped_data["fare"].sum()
total_fare_type

# Build Pie Chart
colors = ["Gold","skyblue","coral"]
explode = (0,0,0.1)
labels = ["Rural","Suburban","Urban"]

plt.pie(total_fare_type,explode=explode, labels=labels,colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)

plt.title("% of Total Fares by City type")
plt.axis("equal")

# Save Figure
plt.savefig("images/total_fare_type.png")
```

```python
# Show Figure
plt.show()
```
